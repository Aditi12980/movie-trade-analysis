```python
import pandas as pd

df = pd.read_csv("movies.csv")
df.head()
```

```python
df.info()
```

```python
#Basic Analysis
print("Total Movies:", len(df))
print(df['Genre'].value_counts())
print(df['Platform'].value_counts())
```

```python
#Visualization
#Genra Distributation
import matplotlib.pyplot as plt

df['Genre'].value_counts().plot(kind='bar')
plt.title("Genre Distribution")
plt.show()
```

```python
#By Seaborn
import seaborn as sns

sns.countplot(data=df, x='Platform')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top rated Movies
top_movies = df.sort_values(by='Rating', ascending=False).head(10)
print(top_movies[['Name', 'Rating']])
```

```python
df['Platform'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Platform Distribution")
plt.ylabel("")
plt.show()
```

```python
df['Month'].value_counts().sort_index().plot(kind='bar')
plt.title("Movies per Month")
plt.xlabel("Month")
plt.ylabel("Number of Movies")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(data=df, x='Genre', y='Rating')
plt.xticks(rotation=45)
plt.title("Genre vs Rating")
plt.show()
```

```python
top_genres = df['Genre'].value_counts().head(3)

top_genres.plot(kind='bar')
plt.title("Top 3 Genres")
plt.show()
```

```python
def rating_category(r):
    if r >= 8:
        return "Excellent"
    elif r >= 6:
        return "Good"
    else:
        return "Average"

df['Rating Category'] = df['Rating'].apply(rating_category)

print(df['Rating Category'].value_counts())
```
